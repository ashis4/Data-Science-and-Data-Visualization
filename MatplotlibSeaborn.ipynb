```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
student_data = pd.read_csv("StudentsPerformance.csv")
```

```python
student_data.head()
```

```python
average_score = (student_data["math score"] + 
                 student_data["reading score"] + 
                 student_data["writing score"]) / 3
```

```python
print(average_score)
```

```python
plt.figure(figsize=(8, 5))
plt.xticks(rotation=45, ha='right', fontsize=10)

plt.bar(student_data["parental level of education"], student_data["reading score"])

plt.tight_layout()
plt.show()
```

```python
plt.hist(average_score, color="orange", edgecolor="red")
plt.title("histogram of average scores")
plt.xlabel("Average scores of students")
plt.ylabel("Number of students")
plt.show()
```

```python
sns.scatterplot(
    data = student_data,
    x = "math score",
    y = "reading score",
    hue = "gender"
)
plt.show()
```

```python

counts = student_data["parental level of education"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title("parental level of education")
plt.show()
```
