```python
math_score = df['math score'].to_numpy()
writing_score = df['writing score'].to_numpy()
reading_score = df['reading score'].to_numpy()

```

```python
students_above_80 = np.sum(
    (math_score > 80) & (writing_score > 80)
)

students_above_80
```

```python
avg_by_gender = df.groupby('gender')['math score'].mean()
avg_by_gender
```

```python
df['average_score'] = (
    df['math score'] +
    df['reading score'] +
    df['writing score']
) / 3
```

```python
avg_score_by_prep = df.groupby('test preparation course')['average_score'].mean()

avg_score_by_prep
```
